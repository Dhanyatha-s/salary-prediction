```python
x=df.iloc[:,[0]].values
y=df.iloc[:,1].values
#x is taken into the 2D always for sklearn and y is always 1D
```

```python

from sklearn.linear_model import LinearRegression as LR
# another way to import library
# import sklearn.linear_model.LinearRegression as LR
```

```python
model= LR()
model.fit(x,y)# create the model fit(mop the i/p and o/p in order)
```

```python
y_pred=model.predict(x)
```

```python
plt.plot(x,y_pred)#line graph best fit line  [0x7f528fdae810>(memory location where the values are stored and predicting graph)]
plt.show()
```

```python

plt.figure(figsize=(8,7))
plt.scatter(x,y,c="k",marker="^",s=50,alpha=0.5,label="actual values")# k for black color , marker is parameter specify  the type of dot we want
plt.plot(x,y_pred,c="#fc8403",label="predicted values")# x-code from color picker
#alpha to malke dots/scatter points to transperant/trancperancy
#s is for make sixze of dots large or small
plt.xlabel("years of experience")
plt.ylabel("execpected salary")
plt.title("Salary vs Age")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(mean_squared_error(y,y_pred))
```
